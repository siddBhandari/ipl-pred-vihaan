```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
import random
```

```python
url = "http://www.howstat.com/cricket/Statistics/Players/PlayerOverviewSummary.asp?PlayerID="
```

```python
data = pd.read_csv("guessplayer.csv")
```

```python
for i in range(len(data)):
    player_id = data["howstatid"][i]
    req_url = url + str(player_id)
    print(i, req_url)
    r = requests.get(req_url)
    soup = BeautifulSoup(r.content, "html.parser")
    try:
        name = soup.find('td', {'class': 'FieldName2'}).find_next('td').text.strip()
    except:
        name = "NA"
        print("----------------------- NA", i)
    data["realname"][i] = name

    time.sleep(random.randint(1, 15))

    print(i, name, data["name"][i])
```

```python
data.to_csv('realnames.csv')
```
