```python
import pandas as pd
```

```python
players = pd.read_csv('player_id.csv')
```

```python
all_players = {}

for i in range(len(players)):
    country_exist = all_players.get(players['country'][i], None)
    if country_exist is None:
        all_players[players['country'][i]] = []
        
    all_players[players['country'][i]].append({
        'name': players['name'][i],
        'howstatId': players['howstatid'][i],
        'iccId': players['iccid'][i],
        'role': players['role'][i],
    })
        
```
