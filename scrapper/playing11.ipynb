```python
import pandas as pd
import json
from fuzzywuzzy import fuzz
```

```python
playing11 = pd.read_csv('playing11.csv')
```

```python
with open("./howstat_json/icc_howstat2.json", "r") as outfile:
    player_json_data = json.loads(outfile.read())
    print(len(player_json_data))
```

```python
country_mapping = {
    "AFG": "afghanistan",
    "AUS": "australia",
    "BAN": "bangladesh",
    "ENG": "england",
    "IND": "india",
    "IRE": "ireland",
    "NZ": "new zealand",
    "PAK": "pakistan",
    "SA": "south africa",
    "SL": "sri lanka",
    "WI": "west indies",
    "ZIM": "zimbabwe",
    "NED": "netherlands",
}
```

```python
for i in range(len(playing11)):
    print(i)
    bowler = False
    batsman = False
    allrounder = False

    for j in range(len(player_json_data)):

        if country_mapping[playing11['country'][i]] == player_json_data[j].get('country', None) and fuzz.partial_ratio(playing11['name'][i].lower(), player_json_data[j]['name'].lower()) > 80:

            playing11['howstatid'][i] = player_json_data[j].get('playerId', -1)
            playing11['iccid'][i] = player_json_data[j].get('iccId', -1)

            fifty = max(int(player_json_data[j]['odi']['batting'].get('50s', 0)), int(player_json_data[j]['t20']['batting'].get('50s', 0)), int(player_json_data[j]['test']['batting'].get('50s', 0)))

            wicket = max(int(player_json_data[j]['odi']['bowling'].get('wickets', 0)), int(player_json_data[j]['t20']['bowling'].get('wickets', 0)), int(player_json_data[j]['test']['bowling'].get('wickets', 0)))

            print(fifty, wicket)

            if fifty >= 5:
                batsman = True
            
            if wicket >= 15:
                bowler = True

            if batsman and bowler:
                playing11['role'][i] = 'allrounder'
            elif batsman:
                playing11['role'][i] = 'batsman'
            elif bowler:
                playing11['role'][i] = 'bowler'
            else:
                playing11['role'][i] = 'unknown'
            # except:
            #     playing11['role'][i] = 'unknown'

            break
```

```python
playing11.to_csv('playing11_populated7.csv', index=False)
```
