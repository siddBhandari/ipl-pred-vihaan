```python
import pandas as pd
import json
```

```python
with open("batting-2024.json","r") as file:
    batting_data = json.load(file)

with open("bowling-2024.json","r") as file:
    bowling_data = json.load(file)

batting_df = pd.DataFrame(batting_data)
bowling_df = pd.DataFrame(bowling_data)
```

```python
batting_df.head()
```

```python
print(len(batting_df),len(bowling_df))
```

```python
bowling_df.head()
```

```python
combined_df = batting_df.merge(bowling_df, on=["ClientPlayerID"], suffixes=('_batting', '_bowling'), how="outer")
combined_df.head()
```

```python
combined_df.columns
```

```python
combined_df.to_json(path_or_buf="combined-2024.json", orient="records")
```
